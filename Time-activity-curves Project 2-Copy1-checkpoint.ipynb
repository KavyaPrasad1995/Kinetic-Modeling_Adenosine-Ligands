```python
Output = pd.read_excel('Raclopride.xlsx',sheet_name='Striatum')
```

```python
cx = sns.boxplot(x='Subject',y='k3/k4',data = Output, width = 0.5, hue = 'Subject',palette = {'Vehicle':'darkgray','CGS21680 1mg/kg':'red','KW6002 1mg/kg':'blue'})
cx.set(xlabel='Group')
cx.set(ylabel='$BP_{ND}$ k3/k4 (2TCM)')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.4, 1),
           borderaxespad=0)
plt.savefig('BPk3k4.png')
```

```python
import seaborn as sns
from statannot import add_stat_annotation
```

```python
cx = sns.boxplot(x='Subject',y='k3/k4',data = Output, hue = 'Subject',palette = {'Vehicle':'darkgray','CGS21680 1mg/kg':'red','KW6002 1mg/kg':'blue'})
cx.set(xlabel='Group')
cx.set(ylabel='$BP_{ND}$ k3/k4 (2TCM)')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.4, 1),
           borderaxespad=0)
add_stat_annotation(cx, x='Subject',y='k3/k4',data = Output, box_pairs=[('Vehicle','CGS21680 1mg/kg'), ('Vehicle','KW6002 1mg/kg')], test = 't-test_ind', text_format = 'star',loc='inside', verbose=2)
plt.savefig('BPk3k4.png')
```

```python
cx = sns.boxplot(x='Subject',y='BPnd',data = Output, hue = 'Subject',palette = {'Vehicle':'darkgray','CGS21680 1mg/kg':'red','KW6002 1mg/kg':'blue'})
cx.set(xlabel='Group')
cx.set(ylabel='$BP_{ND}$ (SRTM)')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.4, 1),
           borderaxespad=0)
add_stat_annotation(cx, x='Subject',y='BPnd',data = Output, box_pairs=[('Vehicle','CGS21680 1mg/kg'), ('Vehicle','KW6002 1mg/kg')], test = 't-test_ind', text_format = 'star',loc='inside', verbose=2)
plt.savefig('SRTM.png')
```

```python
dx = sns.boxplot(x='Subject',y='K1/k2',data = combined, width = 0.6 , hue = 'Region')
dx.set(xlabel='Group')
dx.set(ylabel='$K_{1}/k_{2}$ (2TCM)')
plt.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.4, 1),
           borderaxespad=0)
add_stat_annotation(dx, x='Subject',y='K1/k2',hue = 'Region', data = combined, box_pairs=[(('Vehicle','Cerebellum'),('CGS21680 1mg/kg','Cerebellum')),(('Vehicle','Striatum'),('CGS21680 1mg/kg','Striatum')), (('Vehicle','Striatum'),('KW6002 1mg/kg','Striatum')), (('Vehicle', 'Cerebellum'), ('KW6002 1mg/kg', 'Cerebellum'))] ,test = 't-test_ind', text_format = 'star',loc='inside', verbose=2)
add_stat_annotation(dx, x='Subject',y='K1/k2',hue = 'Region', data = combined, box_pairs=[(('Vehicle','Cerebellum'),('Vehicle','Striatum')),(('CGS21680 1mg/kg','Striatum'),('CGS21680 1mg/kg','Cerebellum')),(('KW6002 1mg/kg','Striatum'),('KW6002 1mg/kg', 'Cerebellum'))] ,test = 't-test_paired', text_format = 'simple',loc='outside', verbose=2)
plt.savefig('K1k2.png')
#Expected one of ['full', 'simple', 'star'] for argument `text_format`; got hash instead.
#test value should be one of the following: t-test_ind, t-test_welch, t-test_paired, Mann-Whitney, Mann-Whitney-gt, Mann-Whitney-ls, Levene, Wilcoxon, Kruskal.
```

```python
dx = sns.boxplot(x='Subject',y='K1/k2',data = combined, width = .6, hue = 'Region', palette = {'Striatum': 'grey', 'Cerebellum':'red'})
dx.set(xlabel='Group')
dx.set(ylabel='$K_{1}/k_{2}$ (2TCM)')
plt.show()
plt.savefig('K1k2.png')
```
