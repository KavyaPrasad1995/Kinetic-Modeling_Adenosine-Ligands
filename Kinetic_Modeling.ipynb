```python
def model_func(x_1,b):
    return ((100 - 7.7) * np.exp(-b*x_1) + 7.7)
```

```python
def model_func_3(x_3,b):
    return ((100 - 9.5) * np.exp(-b*x_3) + 9.5)
```

```python
x_1 = df['Time(Min)']
y_1 = df['Control']
```

```python
x_2 = df['Time(Min)']
y_2 = df['CGS21680']
```
